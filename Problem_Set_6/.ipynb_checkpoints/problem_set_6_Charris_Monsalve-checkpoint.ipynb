```python
#libraries and function that will be used
import cv2
import numpy as np
from matplotlib import pyplot as plt
import os
import pandas as pd
#functions
def PreProcessing(img_resize):
    img=cv2.blur(img_resize.copy(),(5,5))
    out = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV,255,2)
    #_,out=cv2.threshold(img_resize,127,255,cv2.THRESH_BINARY)
    out = cv2.medianBlur(out,3)
    struc = np.ones((3,3),np.uint8)
    #bin_img = cv2.dilate(out, struc, iterations = 1)
    bin_img=out.copy()
    im2, contours, hierarchy = cv2.findContours(bin_img.copy(),cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
    final_img=cv2.drawContours(bin_img, contours, -1, (255,255,255), -1)
    return final_img

def plot_confusion_matrix(df_confusion, title, cmap=plt.cm.gray_r):
    plt.imshow(df_confusion, cmap=cmap) # imshow
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(df_confusion.columns))
    plt.xticks(tick_marks, ['Grass', 'Fire', 'Ghost', 'Water', 'Electric'], rotation=45)
    plt.yticks(tick_marks, ['Grass', 'Fire', 'Ghost', 'Water', 'Electric'])
    for i in range(0,df_confusion.shape[0]):
        for j in range (0,df_confusion.shape[1]):
            plt.text(j, i, "%s%s" % (round(df_confusion[i][j],1),'%'),horizontalalignment="center",color="black" if df_confusion[i][j] < 60 else 'white', fontsize=10, fontweight='bold')
    plt.tight_layout()
    plt.ylabel(df_confusion.index.name)
    plt.xlabel(df_confusion.columns.name)
    plt.show()
```

### Using Hu moments

```python
def Save_Images():
    #Step 1. Resize all images and saved in two folders Preprocessing (Treshold)
    for filename in os.listdir('/home/usuario/Documentos/X Semestre/PDI/Problem_Set_6/PKMN Trainer/'):
        #Reading Images
        img = cv2.imread(os.path.join('/home/usuario/Documentos/X Semestre/PDI/Problem_Set_6/PKMN Trainer/',filename),
                         cv2.IMREAD_GRAYSCALE)
        #plt.figure(figsize=(20,20))
        if filename.endswith('8.jpg') or filename.endswith('9.jpg') or  filename.endswith('10.jpg'):
            img_resize=cv2.resize(img.copy(), (0,0), fx=200/img.shape[1], fy=200/img.shape[0]) 
            final_img=PreProcessing(img_resize)
            cv2.imwrite('/home/usuario/Documentos/X Semestre/PDI/Problem_Set_6/testing_set/%s' % (filename),final_img)
            #Deleting original files
            os.remove('/home/usuario/Documentos/X Semestre/PDI/Problem_Set_6/PKMN Trainer/%s' % (filename))
        else:
            img_resize=cv2.resize(img.copy(), (0,0),fx=200/img.shape[1], fy=200/img.shape[0])
            final_img=PreProcessing(img_resize)
            cv2.imwrite('/home/usuario/Documentos/X Semestre/PDI/Problem_Set_6/training_set/%s' % (filename),final_img)
            #Deleting original files
            os.remove('/home/usuario/Documentos/X Semestre/PDI/Problem_Set_6/PKMN Trainer/%s' % (filename))
```

```python
def Compute_descriptors():    
    #Step 2 Compute descriptors
    #Using SIFT because is a very powerful descriptor.
    #Creating training and response matrix
    i=0
    response=[]
    train_data=[]
    for filename in os.listdir('/home/usuario/Documentos/X Semestre/PDI/Problem_Set_6/training_set/'):
        img = cv2.imread(os.path.join('/home/usuario/Documentos/X Semestre/PDI/Problem_Set_6/training_set/',filename),
                         cv2.IMREAD_GRAYSCALE)
        #FEATURE EXTRACTION
        # Hu moments Scale, translation and rotation invariant
        M = cv2.moments(img)
        des = cv2.HuMoments(M)
        des=des.flatten()

        train_data.append(des)

        if filename.startswith('Vic'):
            response.append(0)
        elif filename.startswith('Vul'):
            response.append(1)
        elif filename.startswith('Gen'):
               response.append(2)
        elif filename.startswith('Gold'):
            response.append(3)
        else:
            response.append(4)
        #print(bag_of_features)
        i=i+1

    train_data = np.array(train_data, dtype='float32')
    response = np.array(response, dtype='float32')
    return train_data, response
```

```python
def Create_KNN(train_data,response):
    # Step 3 Create K-NN clasifier
    knn = cv2.ml.KNearest_create()
    #print(train_data.shape)
    knn.train(train_data,cv2.ml.ROW_SAMPLE,response)
    #ret, results, neighbours ,dist = knn.find_nearest(newcomer, 3)
    return knn
```

```python
def Testing(knn,k):
    #step 4 testing

    actual=np.zeros((15,1))
    classifier=np.zeros((15,1))
    i=0
    for filename in os.listdir('/home/usuario/Documentos/X Semestre/PDI/Problem_Set_6/testing_set/'):
        img_test = cv2.imread(os.path.join('/home/usuario/Documentos/X Semestre/PDI/Problem_Set_6/testing_set/',filename),
                             cv2.IMREAD_GRAYSCALE)
        Hu = cv2.moments(img_test)
        des = cv2.HuMoments(Hu)
        test_des=des.flatten()
        test_des=test_des.reshape(1,7)
        ret, results, neighbours ,dist = knn.findNearest(test_des.astype(np.float32),k)
         #getting real values

        if filename.startswith('Vic'):
            actual[i]=0
        elif filename.startswith('Vul'):
            actual[i]=1
        elif filename.startswith('Gen'):
            actual[i]=2
        elif filename.startswith('Gold'):
            actual[i]=3
        else:
            actual[i]=4

        classifier[i]=results   
        #K-nn results
        if results==0:
            pkmn_type='Grass'
        elif results==1:
            pkmn_type='Fire'
        elif results==2:
            pkmn_type='Ghost'
        elif results==3:
            pkmn_type='Water'
        else:
            pkmn_type='Electric'
        i=i+1
        #print('%s clasified as: %s' % (filename,pkmn_type))
        #print('______________________________________________')
    #Confusión matrix
    y_actu = pd.Series(actual.reshape(15), name='Actual')
    y_pred = pd.Series(classifier.reshape(15), name='Predicted')
    df_confusion = pd.crosstab(y_actu, y_pred,rownames=['True'], colnames=['Predicted'], margins=False)
    df_conf_norm = df_confusion / df_confusion.sum(axis=1)*100
    plot_confusion_matrix(df_conf_norm,'Confusion Matrix k=%s' %(k))        
        
```

```python
#Main Code
def main():
    Save_Images()
    train_data,response=Compute_descriptors()
    knn=Create_KNN(train_data,response)
    for k in range(1,7):
        Testing(knn,k)
        
#using a function that complete the problem set 6        
main()
```

### Implementación:
Para la solución de este problema, se utilizaron principios de funciones utilizadas en problemas desarrollados anteriormente. Se inicia cargando las imagenes ubicadas en la carpeta PKMN Trainer, para luego ser segmentadas. Esto es, separar la figura de los pokemones del fondo. Una vez se obtiene la forma del pokemon en una imagen binaria, 35 son almacenadas en la carpeta training_set y 3 en la carpeta testing_set. Así, se almacenan 7 imagenes por cada tipo de pokemon (7*5=35) en la carpeta training_set y 3 imágenes por tipo (3*5=15) en la carpeta testing_set.

Para entrenar el sistema se utilizan las 35 imagenes que fuero preprocesadas. Utilizando <tt>cv2.HuMoments</tt> se obtienen 7 momentos de hu por imagen, estos vectores se almacenan y apilan en la matrix <tt> train_data </tt>, el vector <tt> response </tt> es utilizado para indicar a que tipo de pokemon pertenece cada una de los 35 momentos de Hu calculados. Luego, se utiliza <tt>knn.train(train_data,cv2.ml.ROW_SAMPLE,response)</tt> para crear un sistema que permita identificar a que tipo pertence un pokemon de la carpeta testing_set.

Se realizaron las pruebas con las imágenes de la carpeta testing_set. Los resultados se muestran en las matrices de confusión. 

### Resultados:
Las matrices de confusión nos muestran un buen desempeño para el sistema. Se observa que a medida que se aumentan el número de vecinos cercanos, la precisión del sistema mejora. Por ejemplo para k=1 se obtiene un pocentaje de exito del 100% solo para el tipo eléctrico. Por otro lado con K=6 se presentan porcentajes de 100% de éxito para los tipos planta, agua y eléctrico. 

Es importante decir que la precisión del sistema puede mejorar si se utiliza un conjunto de datos de entrenamiento más grande. Se utilizaron las cantidades mínimas recomendadas por el dataset, obteniendo buenos resultados al probar con 15 imágenes.

Los 7 momentos de Hu fueron elegidos como descriptores debido son invariantes ante cambios de scala, rotación y traslación. Por lo anterior, este descriptor es adecuado para identificar pokemones, debido a que las imágenes encontradas, presentan cambios de escala, posición y rotación. Para poder calcular estos descriptores es necesario primero crear una imagen binaria con la silueta del pokemon, por eso se utiliza un prerpocesamiento que incluye, umbralización, filtros, detección de bordes y dilatación.
